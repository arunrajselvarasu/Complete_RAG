## Semantic Chunk

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
#loading the embedding model
model = SentenceTransformer('all-MiniLM-L6-V2')
```

```python
#sample text
text = """BMW is a German luxury automobile manufacturer founded in 1916.

BMW is headquartered in Munich, Germany.

BMW produces premium cars, motorcycles, and electric vehicles.

The BMW i4 and BMW iX are popular electric vehicle models.

BMW continues to invest heavily in sustainable transportation technologies.

Mercedes-Benz is another German automotive company known for luxury vehicles.

Mercedes-Benz develops advanced driver assistance and autonomous driving systems.

The Mercedes EQS is a flagship electric luxury sedan.

Audi is a premium automobile manufacturer owned by the Volkswagen Group.

Audi focuses on innovation, performance, and electric mobility.

The Audi e-tron series represents the company's electric vehicle lineup.

Tesla is an American electric vehicle company founded by Elon Musk.

Tesla is known for its Model S, Model 3, Model X, and Model Y vehicles.

Tesla pioneered large-scale adoption of electric vehicles worldwide.

Groq is a technology company specializing in AI inference hardware.

Groq provides ultra-fast inference for large language models.

Groq hardware is designed to reduce latency in AI applications.

LangChain is an open-source framework for building LLM-powered applications.

LangChain provides tools for chains, agents, memory, and retrieval systems.

LangChain is widely used for Retrieval-Augmented Generation (RAG) applications.

Retrieval-Augmented Generation (RAG) combines information retrieval with text generation.

RAG helps language models answer questions using external knowledge sources.

RAG systems often use vector databases to store document embeddings.

ChromaDB is a vector database used for semantic search and retrieval.

ChromaDB stores embeddings and retrieves relevant documents efficiently.

FAISS is another popular vector database library developed by Meta.

Sentence Transformers generate dense vector embeddings for semantic search tasks.

Embeddings convert text into numerical representations that capture meaning.

Cosine similarity measures how similar two embeddings are.

Semantic chunking groups related sentences based on embedding similarity.

Machine learning enables computers to learn patterns from data.

Deep learning is a subset of machine learning based on neural networks.

Large Language Models are trained on massive amounts of text data.

Artificial intelligence is transforming industries such as healthcare, finance, and education."""
```

```python
#split into sentence

sentences = [s.strip() for s in text.split("\n") if s.strip()]
sentences
```

```python
##embeded each of the sentence
embeddings = model.encode(sentences)
embeddings
```

```python
##initialize parameter

threshold = 0.4
chunk = []
current_chunk = [sentences[0]]
```

```python
#semntic group based on threshold

for i in range(1,len(sentences)):
    sim = cosine_similarity(
        [embeddings[i-1]],
        [embeddings[i]]
    )[0][0]

    if sim>=threshold:
        current_chunk.append(sentences[i])
    else:
        chunk.append(" ".join(current_chunk))
        current_chunk=[sentences[i]]
#append the last chunk
chunk.append(" ".join(current_chunk))
```

```python
#see the op of the chunks

print("The chunks \n")
for idx, chun in enumerate(chunk):
    print(f"\nThe chunk {idx+1} : \n{chun}")
```

# Semntic Chunks With Lancchain

```python
from langchain_groq import ChatGroq
from langchain_experimental.text_splitter import SemanticChunker
from langchain.document_loaders import TextLoader
```

```python
loader = TextLoader("Sample.txt")
docs = loader.load()
docs
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
embedding = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
embedding
```

```python
chunker = SemanticChunker(embedding)
```

```python
chunks = chunker.split_documents(docs)
chunks
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
llm = ChatGroq(
    model_name="qwen/qwen3-32b"
)
txt = "tell about chennai"
sam_res=llm.invoke(txt)
sam_res
```

```python
from langchain.vectorstores import FAISS

vectorestores = FAISS.from_documents(chunks,embedding)

retriever = vectorestores.as_retriever()

retriever
```

```python
from langchain.prompts import ChatPromptTemplate
template = {"""Answer the follwing question based on the context:
            {context}
            
            Question:
            {question}"""}
prompt = ChatPromptTemplate.from_messages(template)
prompt
```

```python
from langchain_core.runnables import RunnablePassthrough, RunnableMap
from langchain_core.output_parsers import StrOutputParser

rag_chain = ( 
    RunnableMap(
    {
        "context": lambda x: retriever.invoke(x["question"]),
        "question": lambda x: x["question"],
    }
)
| prompt
| llm
| StrOutputParser()

)
```

```python
response = {"question": "What is biriyani ?"}
res = rag_chain.invoke(response)
res
```
