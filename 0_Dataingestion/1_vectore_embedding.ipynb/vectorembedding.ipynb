## Embedding

```python
import numpy as np
```

```python
def cosine_similarity(vec1,vec2):
    dot_pro=np.dot(vec1,vec2)
    norm_a = np.linalg.norm(vec1)
    norm_b = np.linalg.norm(vec2)
    return dot_pro/(norm_a*norm_b)
    
```

```python
car = [0.82, 0.14, 0.33, 0.91, 0.27]

bike = [0.79, 0.16, 0.30, 0.88, 0.25]

apple = [0.11, 0.92, 0.44, 0.21, 0.77]

banana = [0.13, 0.89, 0.41, 0.19, 0.74]

laptop = [0.66, 0.35, 0.84, 0.55, 0.92]

phone = [0.64, 0.37, 0.81, 0.58, 0.90]

cat = [0.23, -0.52, 0.87, -0.11, 0.45]

kitten = [0.22, -0.49, 0.86, -0.10, 0.44]
```

```python
bike_car_sim = cosine_similarity(car,bike)
print(bike_car_sim)
```

```python
car_banana_sim = cosine_similarity(car,banana)
print(car_banana_sim)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embedding = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
embedding
```

```python
sentences = [
    "The cat is sleeping on the sofa.",
    "The cat is sleeping on the sofa.",
    
    "I bought a new laptop yesterday.",
    "This computer has a powerful processor.",
    
    "The car is parked outside the house.",
    "A bike is standing near the gate.",
    
    "Apple is a healthy fruit.",
    "Banana contains a lot of potassium.",
    
    "Machine learning is used in AI applications.",
    "Artificial intelligence is transforming technology."
]

embeddings_sentence = embedding.embed_documents(sentences)
print(embeddings_sentence[0])
print(embeddings_sentence[1])
print(len(embeddings_sentence[0]))
```

```python
for i in range(len(embeddings_sentence)):
    for j in range(len(embeddings_sentence)):
        similarity=cosine_similarity(embeddings_sentence[i],embeddings_sentence[j])

        print(f"{sentences[i]} vs {sentences[j]}")
        print(f"Similarity : {similarity: .3f}\n")
```


## Embedding with document 

```python
documents = [
    "Python is a popular programming language used for AI and web development.",
    "Machine learning helps computers learn patterns from data.",
    "FAISS is used for vector similarity search in RAG applications.",
    "Groq provides ultra-fast inference for large language models.",
    "LangChain helps developers build LLM-powered applications."
]

query = "Which tool is used for similarity search?"
```

```python
print(documents)
```

```python
def semantic_search(query,documents,embedding_models,top_k=3):

    #embedding for query and document

    query_embeded = embedding_models.embed_query(query)
    document_embeded = embedding_models.embed_documents(documents)

    #calculate similarities

    similarities=[]

    for i, emd_doc in enumerate(document_embeded):
        similarity_score=cosine_similarity(query_embeded,emd_doc)
        similarities.append((similarity_score,documents[i]))

    #sort the embedding

    similarities.sort(reverse=True)
    
    return(similarities[:top_k])
```

```python
result = semantic_search(query,documents,embedding)
result

```
