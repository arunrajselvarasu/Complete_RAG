```python
from langchain.document_loaders import TextLoader

loader = TextLoader("data/textfiles/python_intro.txt",encoding="utf-8")
document=loader.load()
print(document[0].page_content)
print(type(document))
```

```python
from langchain_community.document_loaders import DirectoryLoader

dir_loader = DirectoryLoader(
    "data/textfiles",
    glob="**/*.txt",
    loader_cls= TextLoader,
    loader_kwargs= {'encoding': 'utf=8'},
    show_progress= True
)
document=dir_loader.load()
for i, docs in enumerate(document):
    print(f"Document {i+1}")
    print(f"source {docs.metadata['source']}")
    print(f"length {len(docs.page_content)} Characters")

```

```python
text = document[0].page_content
text
```

```python
char_splitter = CharacterTextSplitter(
    separator=" ",
    chunk_size=100,
    chunk_overlap=10,
    length_function=len
)
char_chunk=char_splitter.split_text(text)
print(f"length of the chunk {len(char_chunk)}")
print(f"Chunks are {char_chunk[0][:100]}")
```

```python
#recursive 
```

```python
rec_splitter = RecursiveCharacterTextSplitter(
    separators=[" ",""],
    chunk_size=100,
    chunk_overlap=10,
    length_function=len
)
rec_chunk = rec_splitter.split_text(text)
print(f"numer of chunks {len(rec_chunk)} ")
```
