# Agent Creation

```python
from langchain_groq import ChatGroq
llm = ChatGroq(model_name="qwen/qwen3-32b")
llm.invoke("Arunraj Selvarasu")
```

```python
#import chat tenplate
from langchain_core.prompts import ChatPromptTemplate
```

```python
from langchain_core.prompts import ChatPromptTemplate
#Researcher Agent

research_prompt = ChatPromptTemplate.from_template("""
You are the Researcher Agent.
{topic}
                                                   
Provide:
1. Overview
2. Key Concept
3. Important Details
4. References                                                  

""")

researcher_chain = research_prompt | llm

#Reviewer Agent

reviewer_prompt = ChatPromptTemplate.from_template("""
You are a Reviewer Agent.
Review the following research:
{research}
Check:
- Accuracy
- Missing Information
- Improvements
Provide detailed feedback.

""")

reviewer_chain = reviewer_prompt | llm

#writer agent

writer_prompt = ChatPromptTemplate.from_template("""
You are the Writer Agent:

Research:
{research}
                                                 

Review and feeedback:
{reviwer}
                                                 
Create a professional report.
Use:
- Headings
- Bullet Points
- Proper Structure
""")

writer_chain = writer_prompt | llm

def Orchestrator(topic:str):
    print("\n" + "=" * 60)
    print("RESEARCHER AGENT")
    print("=" * 60)

    research_response = researcher_chain.invoke({"topic": topic})
    research_response.pretty_print()


    print("\n" + "=" * 60)
    print("REVIEWER AGENT")
    print("=" * 60)

    reviewer_response = reviewer_chain.invoke({"research": research_response.content})
    reviewer_response.pretty_print()


    print("\n" + "=" * 60)
    print("WRITER AGENT")
    print("=" * 60)

    writer_response = writer_chain.invoke({"research": research_response.content, "reviwer": reviewer_response.content})
    writer_response.pretty_print()

    return writer_response.content

result = Orchestrator("What is RAG ?")

print("\n" + "=" * 60)
print("FINAL REPORT")
print("=" * 60)

print(result)
```
