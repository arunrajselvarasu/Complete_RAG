```python
import phonenumbers
from phonenumbers import geocoder, carrier, timezone

number = phonenumbers.parse("+919865927848")

# Country/region (from number prefix only — NOT real-time location)
print(geocoder.description_for_number(number, "en"))  # e.g., "India"

# Carrier
print(carrier.name_for_number(number, "en"))  # e.g., "Airtel"

# Timezone
print(timezone.time_zones_for_number(number))  # e.g., ['Asia/Calcutta']
```
